```python
# Real time capture
import cv2
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    res = segment_image.segmentFrame(frame, show_bboxes=True)
    image = res[1]
    
    cv2.imshow('Instance segmentation', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindow()
```
